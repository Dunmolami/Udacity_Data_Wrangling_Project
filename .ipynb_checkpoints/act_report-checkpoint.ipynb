## Insights from WeRateDogs

WeRateDogs is an account on twitter, a source for professional dog rating. WeRateDogs provides rating for the picture of dogs sent to their page by people. The ratings usually have a denominator of 10 while the numerator is usually greater than 10 (e.g 12/10, 15/10). You might want to ask why, it is like that? It is because they are the best!

WeRateDogs has been verified by Twitter which indicate that they are authentic, notable and active. the account has over 9 million followers and has been in existence on Twitter since November 2015.

WeRateDogs gave Udacity access to their downloaded Twitter archive, an archive that contains basic twwet data (tweet ID, timestamp, source, text, etc.) for over 5000 tweets as of August 1, 2017. Additional data were gotten from Twitter API and from image prediction data. The data were gathered, assessed, cleaned, stored, analysed and visualise to gain some insights into the wrangled data.
Let me take you through some interesting insights I gained from the analysed data.

The Two Most Common Dog Breed Predicted

Images of dogs were posted and the breed of dogs were predicted. The prediction was done three times, the 2 most common dog breeds are Labrador retriever and Golden retriever.

Most common name people give dogs
Different people name their dogs differently. the most commmon names on WeRateDogs is Charlie, followed by Oliver and Penny. pretty names right? i love the names too.


The firt Dog Posted on WeRateDogs page
The first dog posted on WeRateDogs page is Holly and it was rated 11/10. The picture of the dog was posted on the 30th of November 2015. The dog has no pet name and the tweet was sent through an Iphone.
